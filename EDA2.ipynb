```python
# Missing Values

features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]

for feature in features_nan:
    print("The feature is {} and missing values in {}% ".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

# handling missing values
 


```python
# Replacing categorical features

features_categorical = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```

```python
features_categorical
```

```python
data = dataset.copy()
def replace_cat_features(dataset,features_categorical):
#     data[features_categorical] = data[features_categorical].fillna('Missing')
      data[features_categorical] = data[features_categorical].fillna(data[features_categorical].mode())
      return data
    
    
```

```python
replace_cat_features(data,features_categorical)
```

```python
numerical_features_nan = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O' and dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
for feature in numerical_features_nan:
    # since there are outliers we are going to replace with median
    median_values = dataset[feature].median()
    # create a new feature to capture nan values
    dataset[feature+ 'nan'] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_values,inplace=True)
```

```python
## temporal variable(date, time variables)
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature+'age of house']= dataset['YrSold'] - dataset[feature]
```
