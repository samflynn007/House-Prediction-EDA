```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# to display all the columns of the dataframes
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv("train.csv")
```

```python
dataset.head()
```

```python
# print shape of dataset with rows and columns
print(dataset.shape)
```

## in Data we will analyze to find the below stuff
   1. missing values
   2. All the numerical variables
   3. distribution of the numerical variables
   4. categorical variables
   5. outliers
   6. relationship between independent and dependent features
   7. correlations


```python
#missing values
dataset.isnull().sum()
```

```python
feature_with_nan = [features for features in dataset.columns if dataset[features].isnull().sum() >= 1]
```

```python
# these are the features with nan values
feature_with_nan
```

```python
#dataset['Electrical'].isnull().mean()
for feature in feature_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean()*100,4),"% missing values")
```

```python
sns.boxplot(dataset['SalePrice'])
```

```python
# let's plot some amazing diagrams

data = dataset.copy()
for feature in feature_with_nan:
   
    #let's make a variable that indicates 1 if the observastion was missing or 0 otherwise
    data[feature] = np.where(data[feature].isnull(),1,0)
    #lets calculate the mean sale price where thr information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

## numerical variables


```python
# list of numerical variables

dataset['SalePrice'].dtype != 'object'
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']
```

```python
print(len(numerical_features))
```

```python
dataset[numerical_features].head()
```

```python
dataset.info()
```

```python
# Temporal variables(E) - Date Time variables
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs Year Sold")
```

```python
## here we will compare the difference between all years feature with sales price
data = dataset.copy()
for feature in year_feature:
    if feature != 'YrSold':
        data[feature] = data['YrSold'] - data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

## categorical variables

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == "O"]
```

```python
len(categorical_features)
```

```python
dataset[categorical_features].head()
```

```python
for feature in categorical_features:
    print("The feature name is {} and the number of categories are {} ".format(feature,len(dataset[feature].unique())))
```

```python
## find the relationship between categorical feature and sales price
data = dataset.copy()
for feature in categorical_features:
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
```
